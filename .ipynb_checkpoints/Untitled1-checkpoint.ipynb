```python
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point, LineString
import pandas as pd
# 创建一个包含起点和终点的Dataframe
df = pd.DataFrame({
    'start_lon': [116.3975, 121.4737, 120.1551],
    'start_lat': [39.9087, 31.2304, 30.2741],
    'end_lon': [121.4737, 120.1551, 113.9438],
    'end_lat': [31.2304, 30.2741, 22.5293]
})

# 将起点和终点的经纬度信息转换为LineString类型的几何图形
geometry = [LineString([(start_lon, start_lat), (end_lon, end_lat)]) 
            for start_lon, start_lat, end_lon, end_lat in zip(df['start_lon'], df['start_lat'], df['end_lon'], df['end_lat'])]

# 创建一个包含几何图形的geodataframe对象
gdf = gpd.GeoDataFrame(geometry=geometry)

ax = gdf.plot(figsize=(10,10))

# 设置坐标轴标签等其他属性
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title('Map of All Paths')

# 显示地图
plt.show()
```
